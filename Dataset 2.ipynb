```python
# drop the first col named Unnamed: 0
df.drop(['country'], axis=1, inplace=True)
```

```python
# drop the last row 
df.drop([df['show_id'].count()], axis=0, inplace=True)
df.info()
```

```python
# reverse the df for better plot visualization 
df = df.reindex(index=df.index[::-1])
df.head()
```

```python
# clean data as to plot graphs using matplotlib
years = df['Year'].to_list()
years = list(map(int, years))


def remove_per(str1):
    num = str1.split('%')[0]
    return float(num)

growth = df['Growth'].to_list()
growth = list(map(remove_per, growth))


def cleanGDP(str1):
    str1 = str1.replace('$', '')
    str1 = str1.replace(',', '')
    str1 = str1.replace('B', '')
    num = float(str1)
    return num

gdp = df['GDP'].to_list()
gdp = list(map(cleanGDP, gdp))


per_capita = df['Per Capita'].to_list()
per_capita = list(map(cleanGDP, per_capita))
```

```python
# plot Year wise Growth 
fig, ax = plt.subplots()
ax.plot(years, growth)
plt.xlabel("Year_release")
plt.ylabel("Numbers")
plt.title("Year wise release")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(years, gdp, edgecolor="white")
plt.xlabel("Year_relase")
plt.ylabel(".")
plt.title(".")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(years, per_capita, edgecolor="white")
plt.xlabel("Director")
plt.ylabel("Cast)")
plt.title(".")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(gdp, per_capita)
plt.xlabel("Cast")
plt.ylabel("Category")
plt.title("GDP vs Per Capita")
plt.show()
```
